```
# Para crear lista de colores 

colores = []
for i in range(len(test["Estado"])):
  
  if (test["Estado"][i]=="ciudad de mexico"):
    colores.append("#156fad")
      
  if (test["Estado"][i]=="mexico"):
    colores.append("#ad2c15")
  
  if (test["Estado"][i]=="jalisco"):
    colores.append("#ad5e15")

  if (test["Estado"][i]=="morelos"):
    colores.append("#15ad40")

  if (test["Estado"][i]=="nuevo leon"):
    colores.append("#15a8ad")

  if (test["Estado"][i]=="oaxaca"):
    colores.append("#839ac9")

  if (test["Estado"][i]=="puebla"):
    colores.append("#bc83c9")

  if (test["Estado"][i]=="tamaulipas"):
    colores.append("#d66598")

  if (test["Estado"][i]=="veracruz de ignacio de la llave"):
    colores.append("#ed9145")
```

```
#Truco para que las etiquetas quedaran bien

aux=test[test["Estado"]==0].head(1)
for i in list(set(test["Estado"])):
  aux = pd.concat([aux,test[test["Estado"]==i].head(1)]) 
aux = aux.reset_index()

colores_aux = []
for i in range(9):
  if (aux["Estado"][i]=="ciudad de mexico"):
    colores_aux.append("#156fad")
      
  if (aux["Estado"][i]=="mexico"):
    colores_aux.append("#ad2c15")
  
  if (aux["Estado"][i]=="jalisco"):
    colores_aux.append("#ad5e15")

  if (aux["Estado"][i]=="morelos"):
    colores_aux.append("#15ad40")

  if (aux["Estado"][i]=="nuevo leon"):
    colores_aux.append("#15a8ad")

  if (aux["Estado"][i]=="oaxaca"):
    colores_aux.append("#839ac9")

  if (aux["Estado"][i]=="puebla"):
    colores_aux.append("#bc83c9")

  if (aux["Estado"][i]=="tamaulipas"):
    colores_aux.append("#d66598")

  if (aux["Estado"][i]=="veracruz de ignacio de la llave"):
    colores_aux.append("#ed9145")
```

```
plt.figure(figsize=(20,10))

for i in range(9):
  plt.scatter(aux["Municipio"][i],aux["Poblacion"][i], label=aux["Estado"][i] ,c=colores_aux[i])

plt.scatter(test["Municipio"],test["Poblacion"], label='Poblacion',c=colores)
plt.legend(loc="upper right",fontsize="large")
plt.xticks(rotation=90,fontsize="large")
plt.show()
```

```
#Para determinar el tamaño del gráfico
plt.figure(figsize=(20,10))

#Para que se vean bonitas las etiquetas
for i in range(9):
  plt.scatter(aux["Municipio"][i],aux["Número de unidades"][i], label=aux["Estado"][i] ,c=colores_aux[i])

#Para graficar

plt.scatter(test["Municipio"],test["Número de unidades"],c=colores)
plt.legend(loc="upper right",fontsize="large")
plt.xticks(rotation=90,fontsize="large")
plt.show()
```
